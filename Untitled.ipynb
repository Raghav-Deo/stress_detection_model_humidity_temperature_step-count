# Imports

```python
import pandas as pd
import numpy as np

from pandas.plotting import scatter_matrix

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler\

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import load_model
```

# Dataset importing

```python
df = pd.read_csv("Stress-Lysis.csv")
```

# Exploratory data analysis

```python
print("Output of df.head(10)\n", df.head(10), "\n")
print("Output of df.info()\n", df.info(), "\n")
print("Output of df.describe()\n", df.describe())
```

# Data Visualisation

```python
df.hist(figsize = (20, 15), bins = 50);
scatter_matrix(df[["Humidity", "Temperature", "Step count", "Stress Level"]], figsize = (20, 15));
```

# Train Test Splitting

```python
train_set, test_set = train_test_split(df, test_size=0.30, random_state=42)
```

# Separating features and labels and standardising data

```python
scaler = StandardScaler()
train_features = scaler.fit_transform(train_set.drop(columns = "Stress Level"))
test_features = scaler.fit_transform(test_set.drop(columns = "Stress Level"))
train_label = train_set["Stress Level"]
test_label = test_set["Stress Level"]
```

# Creating a neural network

```python
model = Sequential()
model.add(Dense(2, activation = "relu"))
model.add(Dense(6, activation = "relu"))
model.add(Dense(3, activation = "softmax"))
model.compile(optimizer = 'adam',loss = 'sparse_categorical_crossentropy',metrics = ['accuracy']);
model.fit(train_features, train_label, epochs = 23, shuffle = True, validation_data = (test_features, test_label));
print(model.summary(), "\n\n")
```

# Saving the model to later load with keras

```python
model.save("stress_detection.h5")
```

```python
temp_model = load_model("stress_detection.h5")
temp_model.summary();
```

```python
result = np.argmax(model.predict(scaler.transform([[21.41, 90.41, 93]])))
print(result)
```
